```python
# Determination of the error in U0/t0 due to the 
# systematic uncertainty in the lattice depth

from uncertainties import ufloat,unumpy
from scipy import stats

wF = np.loadtxt('../EXP/tANDU.dat')

v0 = wF[:,0]
t  = wF[:,1]
wFac = wF[:,2]

aS = 290. 
a0a = 5.29e-11 / 532e-9

U = aS*a0a * wFac / t

plt.plot( v0, U, '.-' )

ax = plt.gca()
ax.set_xlabel('$v_{0}$')
ax.set_ylabel('$U/t$')
ax.set_xlim(6.,8.)
ax.set_ylim(8., 15.)


subset = np.logical_and( v0>6.5, v0<7.5)

z = np.polyfit( v0[subset], U[subset] , 1)  
p = np.poly1d(z) 

xpts = np.linspace(6.5, 7.5, 50)
ypts = p(xpts)
ax.plot( xpts, ypts, 'k' )

print p( 7.  )

#
# get the slope of wF^(1/3)  in unitsof wF^(1/3) 
slope_wF13 = z[0] / p(7.) 

print slope_wF13

plt.show()
```

```python
# Determination of the error in U0/t0 due to the 
# systematic uncertainty in the lattice depth

from uncertainties import ufloat,unumpy
from scipy import stats

wF = np.loadtxt('../EXP/tANDU.dat')
wF = wF[ np.logical_and(  wF[ :, 0] > 6.5 , wF[:,0]<7.5  ) ] 
#
z = np.polyfit( wF[:,0], np.power( wF[:,2], 1./3.) , 1)  
p_wF = np.poly1d(z) 
x = np.linspace( 6., 8., 20) 
plt.plot( wF[:,0], np.power(wF[:,2], 1./3.), '.', color='black')
plt.plot( x, p_wF(x), '-' )  
ax = plt.gca()
ax.set_xlabel('$v_{0}$')
ax.set_ylabel('$(\mathrm{wF})^{1/3}$')
plt.show()
#
z_t = np.polyfit( wF[:,0], wF[:,1], 1 )
p_t = np.poly1d(z_t) 
x = np.linspace( 6., 8., 20) 
plt.plot( wF[:,0], wF[:,1], '.', color='black')
plt.plot( x, p_t(x), '-' )  
ax = plt.gca()
ax.set_xlabel('$v_{0}$')
ax.set_ylabel('$t$')
plt.show()




# get the uncertainty in the lattice depth for x,y,z
sx = ufloat( 21.8, 0.3)
sy = ufloat( 21.6, 0.2)
sz = ufloat( 18.6, 0.5)

def get_v0( s ):
    mW = 7.0 * s.nominal_value 
    return  mW / s

v0x = get_v0( sx)
v0y = get_v0( sy)
v0z = get_v0( sz)

v0x= ufloat( 7., 0.05*7.)
v0y= ufloat( 7., 0.05*7.)
v0z= ufloat( 7., 0.05*7.)

print "Lattice depths:"
print v0x
print v0y
print v0z

print
print "wF^(1/3)'s : "
print p_wF( v0x )
print p_wF( v0y )
print p_wF( v0z )


print 
print "t's :"
print p_t( v0x )
print p_t( v0y )
print p_t( v0z )


aS = 290. 
a0a = 5.29e-11 / 532e-9

U = aS*a0a * ( p_wF(v0x) * p_wF(v0y) * p_wF(v0z)   ) / ((  p_t(v0x) + p_t(v0y) + p_t(v0z))/3.)
print 
print "U:"
print U
 
#s0_err =  np.sqrt(  get_err_sq( sx ) \
#             + get_err_sq( sy ) + get_err_sq( sz ) )

#U_err_factor = slope_wF13 * s0_err 

#print slope_wF13
#print get_err_sq( sx ) 
#print get_err_sq( sy ) 
#print get_err_sq( sz ) 
#print  s0_err 
#print U_err_factor
```
