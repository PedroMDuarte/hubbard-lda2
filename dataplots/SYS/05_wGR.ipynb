```python
# Start matplotlib
from matplotlib import rc
rc('text', usetex=True)
plt.rcParams['text.latex.preamble'] = [
       r'\usepackage{bm}',        # for bold math
       r'\usepackage{siunitx}',   # i need upright \micro symbols, but you need...
       r'\sisetup{detect-all}',   # ...this to force siunitx to actually use your fonts
       r'\usepackage[textlf]{MyriadPro}',
       r'\usepackage{sansmath}',  # load up the sansmath so that math -> helvet
       r'\sansmath',               # <- tricky! -- gotta actually tell tex to use!
]    
plt.rcParams['axes.linewidth'] = 0.55
plt.rcParams['patch.linewidth'] = 0.4

gs = matplotlib.gridspec.GridSpec(1,2,\
      wspace=0.75, hspace=0.05,\
      left=0.10, right=0.6, bottom=0.57, top=0.95)

fig = plt.figure( figsize(8.3,4.5))

ax = fig.add_subplot( gs[0] )
ax1 = fig.add_subplot( gs[1] )


ax.axhspan( 1.962-0.115, 1.962+0.115, facecolor='gray', alpha=0.6)
ax1.axhspan( 1.962-0.115, 1.962+0.115, facecolor='gray', alpha=0.6)



ax.set_xlabel(xlabel)
ax.set_ylabel(r'$S_{\bm{\pi}}$', rotation=0, labelpad=8.)
ax1.set_xlabel(r'$T/t_{*}$')
ax1.set_ylabel(r'$S_{\bm{\pi}}$', rotation=0, labelpad=8.)



gsProf = matplotlib.gridspec.GridSpec(1,3,\
            wspace=0.5, hspace=0.05,\
            left=0.085, right=0.975, bottom=0.09, top=0.44)

def smooth(y,win):
    return np.convolve(y, np.ones(win)/win)[win/2-1:-win/2]
#smoothed = smooth(y,15)

axProf = fig.add_subplot( gsProf[0] )
axProf.set_xlim( 0., 30.)
axProf.set_xlabel(r'$r\ (\si{\micro\metre})$')
axProf.set_ylabel(r'$n(r)$')
axProf.axhspan( 1.0, 1.3, facecolor='lightgray', alpha=0.3)

axSProf = fig.add_subplot( gsProf[1] )
axSProf.set_xlim( 0., 30.)
axSProf.set_xlabel(r'$r\ (\si{\micro\metre})$')
axSProf.set_ylabel(r'$s_{\bm{\pi}}(r)$')

axHProf = fig.add_subplot( gsProf[2] )
axHProf.set_xlim( 0., 30.)
axHProf.set_xlabel(r'$r\ (\si{\micro\metre})$')
axHProf.set_ylabel(r'$4\pi r^{2} s_{\bm{\pi}}(r)/N$')


gsU = matplotlib.gridspec.GridSpec(1,1,\
        wspace=0.1, hspace=0.1,\
        left=0.79, right=0.97, bottom=0.57, top=0.95)
axU = fig.add_subplot( gsU[0] )
axU.set_xlabel(xlabel)
axU.set_ylabel(r'$U_{0}/t_{0}$')
axU.set_ylim( 9., 13. )

ax.set_xlim(36., 45.) 
axU.set_xlim(36., 45.)

ax.xaxis.set_major_locator( matplotlib.ticker.MultipleLocator(2.0))
ax.xaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(1.0))
axU.xaxis.set_major_locator( matplotlib.ticker.MultipleLocator(2.0))
axU.xaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(1.0))

ax1.set_xlim(0.36, 0.64) 
ax.set_ylim(1.0, 2.6)
ax1.set_ylim(1.0, 2.6)

#==========================
# SET T-BOUNDING PARAMETERS
#==========================
DOBOUND = True

xlbound = 39.15
xubound = 40.85
Slo = 1.962-0.115
Shi = 1.962+0.115

TminA = 0.35
TminB = 0.42

TmaxA = 0.52
TmaxB = 0.54


def annotate_dim(ax,xyfrom,xyto,text=None):
    if text is None:
        text = str(np.sqrt( (xyfrom[0]-xyto[0])**2 + (xyfrom[1]-xyto[1])**2 ))
    ax.annotate("",xyfrom,xyto,arrowprops=dict(arrowstyle='<->'))
    ax.text((xyto[0]+xyfrom[0])/2,(xyto[1]+xyfrom[1])/2,text,fontsize=16)


import glob 
from colorChooser import rgb_to_hex, cmapCycle
import pickle
import os
from scipy.optimize import brentq

def spi_extrapolate( T, s0, beta):
    TN = 0.36
    return 0.5+ s0*np.abs( T/TN - 1 )**(-2.*beta) 

from scipy.optimize import curve_fit

def fit_spi( xT, yS):
    s0 = 0.7; beta=0.3639
    popt, pcov = curve_fit(spi_extrapolate, xT, yS, p0=[s0,beta])
    xFit = np.linspace(0.361, 0.8, 100)
    def p(x):
        return spi_extrapolate( x, *popt)
    yFit = p(xFit) 
    return xFit, yFit, p

# pck keys are :
# U111
# n111_fine
# entrbulk
# t111
# r111_fine
# spi111_fine
# Number
# Tn
# entr111
# aS
# savedir
# n111
# SpiBulk
# r111
# Tspi
# spi111
# dens111
# ldanum

runname = '05_wGR'
key = 'wGR_' 
xlabel = r'$w_{\mathrm{GR}}$'
x0 = 47./1.175 

ms = 3.5
mew = 0.7 
lw = 0.7
mfc = 'white'
cs = ['b', 'g', 'r', 'c', 'm', 'y', 'k']



Tstars =[] 
xs=[]
for fpck in glob.glob( runname +'/*.pck'):
    name = os.path.basename(fpck)
    x = float(name.split(key)[0])
    Tstar = float(name.split(key)[1].split('T')[0] )
    Tstars.append( Tstar )
    xs.append( x ) 

Tstars = np.unique( Tstars )
xs = np.unique( xs ) 

#======================
# PLOT #0 -- CONSTANT T
#======================
for ii,Ts in enumerate(sorted(Tstars)):
    result = [] 
    for fpck in sorted(glob.glob( runname + '/*.pck')):
        name = os.path.basename(fpck)
        x = float(name.split(key)[0])
        Tstar = float(name.split(key)[1].split('T')[0])
        
        if Tstar == Ts: 
            spis = pickle.load( open( fpck,"rb" ) )
            spi = spis['SpiBulk']
            result.append( [x, spi]) 
            
    result = np.array(result)
    ax.plot( result[:,0], result[:,1], 'o-', ms=ms, mew=mew, lw=lw,\
            mfc=mfc, mec=cs[ii], color=cs[ii],\
            label='{:0.2f}'.format(Ts))
    

    
handles, labels = ax.get_legend_handles_labels()
l0 = legend = ax.legend(handles, labels, title='$T/t_{*}$',bbox_to_anchor=(1.02,0.95), loc='upper left', numpoints=1, \
    prop={'size':9}, handlelength=1.1, handletextpad=0.4, ncol=1, columnspacing=0.7, labelspacing=0.55)
plt.setp( legend.get_title(), fontsize=10.)
ax.add_artist(l0)



ax.tick_params(axis='both', which='major',  length=1.5)
ax.tick_params(axis='both', which='minor',  length=1.0)    



csx = [  cmapCycle( matplotlib.cm.gist_rainbow, float(i), lbound=0.,  ubound=len(xs)-1, inv=False) for i in range(len(xs)) ]
#======================
# PLOT #1 -- CONSTANT x
#======================
Us = [] ; Tmin=None; Tmax=None
for ii,xx in enumerate(sorted(xs)):
    if xx == x0:
        mec_ = 'black'
        color_ = 'black'
        lw_ = lw * 1.5
        dash_ = (3.5,1.3,1.3,1.3)
        #[(2.4,1.4), (3.,1,1,1),(1,0.001),(1,1.05)]
    else:
        mec_ = csx[ii]
        color_ = csx[ii] 
        lw_ = lw
        dash_ = (1,0.001)
        
    result = [] 
    for fpck in sorted(glob.glob( runname + '/*.pck')):
        name = os.path.basename(fpck)
        x = float(name.split(key)[0])
        Tstar = float(name.split(key)[1].split('T')[0])
        
        if x == xx: 
            spis = pickle.load( open( fpck,"rb" ) )
            spi = spis['SpiBulk']
            result.append( [Tstar, spi]) 
            
            if Tstar == 0.46:
                nprof = smooth(spis['n111'],5)
                Sprof = smooth( spis['spi111']*spis['n111'],5)
                Hprof = smooth( (1/1.68e5)*4.*np.pi*(spis['r111']**2.)*spis['spi111']*spis['n111'],5)
                
                axProf.plot(  spis['r111'], nprof, color=color_, lw=lw_, dashes=dash_ )
                axSProf.plot( spis['r111'], Sprof, color=color_, lw=lw_, dashes=dash_ )
                axHProf.plot( spis['r111'], Hprof, color=color_, lw=lw_, dashes=dash_ )
                
                Us.append( spis['U111'].max()  )
        
    result = np.array(result)

    xdat = result[:,0]
    ydat = result[:,1]
    
    try:
        xFit, yFit, p = fit_spi( xdat, ydat)
        ax1.plot( xFit, yFit, ':', color=color_, lw=0.4)
        
        if np.abs(xx-xlbound) <=0.01 and DOBOUND:
            def f(x):
                return p(x)-Shi
            Tmin = brentq(f, TminA, TminB )

        if np.abs(xx-xubound) <=0.01 and DOBOUND:
            def f(x):
                return p(x)-Slo
            Tmax = brentq(f, TmaxA, TmaxB )
        
    except:
        z = np.polyfit( xdat, ydat, 3)
        p = np.poly1d(z) 
        Trange = np.linspace(0.35, 0.65, 50)
        ax1.plot( Trange, p(Trange), '--', color=color_, lw=0.4)
        
        if np.abs(xx-xlbound) <=0.01 and DOBOUND:
            Tmin = brentq(p-Shi, TminA, TminB )

        if np.abs(xx-xubound) <=0.01 and DOBOUND:
            Tmax = brentq(p-Slo, TmaxA, TmaxB )
    
    ax1.plot( xdat,ydat, 'o-', ms=ms, mew=mew, lw=lw,\
             mfc=mfc, mec=mec_, color=color_,\
             label='{:0.2f}'.format(xx))
    
        
axU.plot( sorted(xs), Us, '.-', color='k')
axU.axhspan( 11.1*0.92, 11.1*1.08,  color='lightgray', alpha=0.3)
        
if DOBOUND:
    
    print "Tmin = ", Tmin
    print "Tmax = ", Tmax

    ax1.annotate(
        '', xy=(Tmin, 1.24), xycoords = 'data',
        xytext = (Tmax, 1.245), textcoords = 'data',
        arrowprops = {'arrowstyle':'<|-|>', 'facecolor':'black'})
    ax1.axvline( Tmin, color='black', lw=0.4)
    ax1.axvline( Tmax, color='black', lw=0.4)
    ax1.text( Tmin, 1.0, '${:0.2f}$'.format(Tmin), \
             transform = matplotlib.transforms.blended_transform_factory(
        ax1.transData, ax1.transAxes), \
             fontsize=8., ha='center', va='bottom')
    ax1.text( Tmax, 1.0, '${:0.2f}$'.format(Tmax), \
             transform = matplotlib.transforms.blended_transform_factory(
        ax1.transData, ax1.transAxes), \
             fontsize=8., ha='center', va='bottom')



        
        
    
handles, labels = ax1.get_legend_handles_labels()
l0 = legend = ax1.legend(handles, labels, title=xlabel,bbox_to_anchor=(1.02,0.95), loc='upper left', numpoints=1, \
    prop={'size':9}, handlelength=1.1, handletextpad=0.4, ncol=1, columnspacing=0.7, labelspacing=0.55)
plt.setp( legend.get_title(), fontsize=10.)
ax1.add_artist(l0)


ax1.xaxis.set_major_locator( matplotlib.ticker.MultipleLocator(0.10))
ax1.xaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(0.05))


ax1.tick_params(axis='both', which='major',  length=1.5)
ax1.tick_params(axis='both', which='minor',  length=1.0)    



fig.savefig(runname + '.png', dpi=350 )
    
    
```
