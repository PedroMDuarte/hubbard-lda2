```python
from matplotlib import rc
rc('font', **{'family':'serif'})
rc('text', usetex=True)

import sys
sys.path.append('../')

import os
from kappa import kappa0

from scipy.interpolate import interp1d
tInterp = interp1d( np.loadtxt('../banddat/interpdat_t_v0.dat'), np.loadtxt('../banddat/interpdat_t_tCalc.dat'))
def t(v0):
    return float(tInterp(v0))

emassdat = np.loadtxt('../banddat/interpdat_effective_mass.dat')
emass = interp1d( emassdat[:,0], emassdat[:,1] )
def effMass(v0):
    return float( emass(v0))



gs = matplotlib.gridspec.GridSpec(1,4,\
      wspace=0.05, hspace=0.05,\
      left=0.08, right=0.99, bottom=0.13, top=0.95)



xs = 10.; ys=2.5
fign = plt.figure( figsize(xs,ys))
figSn = plt.figure( figsize(xs,ys))
figSnn = plt.figure( figsize(xs,ys))
figCompr = plt.figure( figsize(xs,ys))
figComprn = plt.figure( figsize(xs,ys))
figComprB = plt.figure( figsize(xs,ys))
figComprBn = plt.figure( figsize(xs,ys))


from colorChooser import rgb_to_hex, cmapCycle



Uvals = sorted(   range(0,32,2) + [1.,7.,9.] ) 
Uvals = [0., 4., 8., 20.]
Tvals = sorted( np.unique([0.40, 0.50, 0.64, 0.72, 0.84, 0.90, 1.00, 1.20, 1.40, 1.60, 1.80, 2.0, 2.2, 2.4] \
                + [0.357, 0.360, 0.370, 0.385, 0.400, 0.500, 0.588, 0.667, 1.000]) )

#Uvals = [16.]
ROW = 0 
for i,Uval in enumerate( Uvals ):

    axn = fign.add_subplot( gs[ROW,i])
    axSn= figSn.add_subplot( gs[ROW,i] )
    axSnn = figSnn.add_subplot( gs[ROW,i] )
    axCompr = figCompr.add_subplot( gs[ROW,i] )
    axComprn = figComprn.add_subplot( gs[ROW,i] )
    axComprB = figComprB.add_subplot( gs[ROW,i] )
    axComprBn = figComprBn.add_subplot( gs[ROW,i] )




    for j, Tval in enumerate(Tvals):
        color = cmapCycle( matplotlib.cm.spectral, float(j), lbound=0, ubound=float(len(Tvals)+0))

        # PLOT THE DATA 
        # data columns 
        MUCOL   = 0
        DENSCOL = 1
        ENTRCOL = 2
        SPInCOL = 3
        CMPRCOL = 4
        CMPBCOL = 5 
        ms = 1.6; mew=0.3

        # get SPI data 
        fname =  '../COMB_Final_Spi/'+ 'U{U:02d}/T{T:0.3f}.dat'.format(U=int(Uval),T=Tval)
        if not os.path.isfile(fname):
            #print "File does not exist: ", fname 
            pass
        else:
            try:
                dat = np.loadtxt(fname)

                for ax,COL in [(axn,DENSCOL), (axSn,SPInCOL) ]:
                    ax.plot( dat[:,MUCOL], dat[:,COL], color=color,\
                            #'o', ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))

                axSnn.plot( dat[:,DENSCOL], dat[:,SPInCOL], color=color,\
                            label='${:0.3f}$'.format(Tval))



                gspin = 2. 
                numfactor = (6.* np.pi**2.)**(2./3.) / (2. * gspin**(2./3.) )
                axComprn.plot( dat[:,DENSCOL], dat[:,CMPRCOL] * numfactor , color=color,\
                            label='${:0.3f}$'.format(Tval))
                axCompr.plot( dat[:,MUCOL], dat[:,CMPRCOL] * numfactor, color=color,\
                            #'o', ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))

                axComprB.plot( dat[:,MUCOL], dat[:,CMPBCOL], color=color,\
                            #'o', ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))
                axComprBn.plot( dat[:,DENSCOL], dat[:,CMPBCOL] , color=color,\
                            label='${:0.3f}$'.format(Tval))


            except Exception as e:
                print 'ERROR on U={U:02d}  T={T:0.3f}'.format(U=int(Uval),T=Tval)
                raise e 
                pass 


    npts = np.linspace(0.01, 3.,120)
    t0 = 2. / (np.pi**2.)
    axComprBn.plot( npts, kappa0(npts,gspin)*t0 , '--', color='black', lw=1.)




    hl = 0.8; nc=2; cs=0.3
    axn.legend(title='$T/t$',bbox_to_anchor=(1.02,-0.02), loc='lower right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axSn.legend(title='$T/t$',bbox_to_anchor=(1.02,0.99), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axSnn.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs )
    axCompr.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axComprn.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axComprB.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axComprBn.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)

    axn.set_ylim(-0.2, 2.2) 
    axSn.set_ylim(-0.2,12.0)
    axSnn.set_ylim(-0.2,12.0)
    axCompr.set_ylim(-0.02,1.5)
    axComprn.set_ylim(-0.02,1.5)
    axComprB.set_ylim(1e-2,20000.0); axComprB.set_yscale('log')
    axComprBn.set_ylim(1e-2,20000.0); axComprBn.set_yscale('log')

    if i != 0:
            axn.yaxis.set_ticklabels([])
            axSn.yaxis.set_ticklabels([])
            axSnn.yaxis.set_ticklabels([])
            axCompr.yaxis.set_ticklabels([])
            axComprn.yaxis.set_ticklabels([])
            axComprB.yaxis.set_ticklabels([])
            axComprBn.yaxis.set_ticklabels([])
    else:
        axn.set_ylabel('$n$', rotation=0,labelpad=11)
        axSn.set_ylabel(r'$\frac{S_{\pi}}{n}$', rotation=0,labelpad=11, fontsize=14)
        axSnn.set_ylabel(r'$\frac{S_{\pi}}{n}$', rotation=0,labelpad=11, fontsize=14)
        axCompr.set_ylabel(r"$\kappa'(n)$", rotation=90,labelpad=8, fontsize=14)
        axComprn.set_ylabel(r"$\kappa'(n)$", rotation=90,labelpad=8, fontsize=14)
        axComprB.set_ylabel(r'$\kappa_{T}(n) \frac{t}{a^{3}}$', rotation=90,labelpad=8, fontsize=14)
        axComprBn.set_ylabel(r'$\kappa_{T}(n) \frac{t}{a^{3}}$', rotation=90,labelpad=8, fontsize=14)


    if ROW < 0 :
        axn.xaxis.set_ticklabels([])
        axSn.xaxis.set_ticklabels([])
        axSnn.xaxis.set_ticklabels([])
        axCompr.xaxis.set_ticklabels([])
        axComprn.xaxis.set_ticklabels([])
        axComprB.xaxis.set_ticklabels([])
        axComprBn.xaxis.set_ticklabels([])
    else:
        axn.set_xlabel(r'$\mu/t$',labelpad=-1, fontsize=14)
        axSn.set_xlabel(r'$\mu/t$',labelpad=-1, fontsize=14)
        axSnn.set_xlabel(r'$n$',labelpad=-1, fontsize=14)
        axCompr.set_xlabel(r'$\mu/t$',labelpad=-1, fontsize=14)
        axComprn.set_xlabel(r'$n$',labelpad=-1, fontsize=14)
        axComprB.set_xlabel(r'$\mu/t$',labelpad=-1, fontsize=14)
        axComprBn.set_xlabel(r'$n$',labelpad=-1, fontsize=14)


    axn.set_xlim(-10.,36)
    axSn.set_xlim(-10.,36)
    axSnn.set_xlim(-0.2,2.2)
    axCompr.set_xlim(-10., 36.)
    axComprn.set_xlim(-0.2, 2.2)
    axComprB.set_xlim(-10., 36.)
    axComprBn.set_xlim(-0.2, 2.2)


    for ax in [axn, axSn, axSnn, axCompr, axComprn]:
        ax.xaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6, prune='lower'))


    toplabel = '$U/t={:d}$'.format(int(Uval)) 

    for ax in [axn, axSn, axSnn, axCompr, axComprn]:
        ax.grid(alpha=0.4)
        ax.text( 0.08, 1.02, toplabel, \
                 ha='left', va='top', transform=ax.transAxes,
                 bbox={'boxstyle':'round','facecolor':'white'})

    for ax in [axComprB, axComprBn]:
        ax.grid(alpha=0.4)
        ax.text( 0.28, 1.02, toplabel, \
                 fontsize=10,\
                 ha='left', va='top', transform=ax.transAxes,
                 bbox={'boxstyle':'round','facecolor':'white'})




suffix = '_ROW.png'

#fign.savefig('KAPPA/density_All'.format(Tval) + suffix, dpi=300)
plt.close(fign)
#figSn.savefig('KAPPA/spi_n_All'.format(Tval) + suffix, dpi=300)
plt.close(figSn)
#figSnn.savefig('KAPPA/spi_n_n_All'.format(Tval) + suffix, dpi=300)
plt.close(figSnn)
#figCompr.savefig('KAPPA/compressibility_All'.format(Tval) + suffix, dpi=300)
plt.close(figCompr)
figComprn.savefig('KAPPA/compressibility_n_All'.format(Tval) + suffix, dpi=300)
plt.close(figComprn)
#figComprB.savefig('KAPPA/compressibilityB_All'.format(Tval) + suffix, dpi=300)
plt.close(figComprB)
figComprBn.savefig('KAPPA/compressibilityB_n_All'.format(Tval) + suffix, dpi=300)
plt.close(figComprBn)
```
