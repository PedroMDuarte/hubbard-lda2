```python
from matplotlib import rc
rc('font', **{'family':'serif'})
rc('text', usetex=True)

import sys
sys.path.append('../')

import os

from kappa import compressibility


gs = matplotlib.gridspec.GridSpec(4,5,\
      wspace=0.05, hspace=0.05,\
      left=0.05, right=0.99, bottom=0.08, top=0.96)

xs = 15.2; ys=9.2
fign = plt.figure( figsize(xs,ys))
figSn = plt.figure( figsize(xs,ys))
figSnn = plt.figure( figsize(xs,ys))
figEntr = plt.figure( figsize(xs,ys))
figCompr = plt.figure( figsize(xs,ys))
figComprn = plt.figure( figsize(xs,ys))

from colorChooser import rgb_to_hex, cmapCycle



Uvals = sorted(   range(0,32,2) + [1.,7.,9.] ) 
Tvals = sorted( np.unique([0.40, 0.50, 0.64, 0.72, 0.84, 0.90, 1.00, 1.20, 1.40, 1.60, 1.80, 2.0, 2.2, 2.4] \
                + [0.357, 0.360, 0.370, 0.385, 0.400, 0.500, 0.588, 0.667, 1.000]) )

#Uvals = [16.]

for i,Uval in enumerate( Uvals ):

    axn = fign.add_subplot( gs[i])
    axSn= figSn.add_subplot( gs[i] )
    axSnn = figSnn.add_subplot( gs[i] )
    axEntr = figEntr.add_subplot( gs[i] )
    axCompr = figCompr.add_subplot( gs[i] )
    axComprn = figComprn.add_subplot( gs[i] )
    
    for j, Tval in enumerate(Tvals):
    
        color = cmapCycle( matplotlib.cm.spectral, float(j), lbound=0, ubound=float(len(Tvals)+0))
        
        # PLOT THE NLCE DATA
        # data columns 
        MUCOL = 0
        DENSCOL = 1
        SPInCOL = 4
        ENTRCOL = 2

        fname =  '../NLCE8_Final/'+ 'U{U:02d}/T{T:0.2f}.dat'.format(U=int(Uval),T=Tval)
        if not os.path.isfile(fname):
            #print "File does not exist: ", fname 
            pass
        else:
            dat = np.loadtxt(fname)

            # NLCE8_Final columns
            # 0  chemical potential
            # 1  density
            # 2  entropy
            # 4  spi/n
            
            try:

                for ax,COL in [(axn,DENSCOL), (axSn,SPInCOL)  ]:
                    ax.plot( dat[:,MUCOL], dat[:,COL], color=color, label='${:0.2f}$'.format(Tval))
                axSnn.plot( dat[:,DENSCOL], dat[:,SPInCOL], color=color, label='${:0.2f}$'.format(Tval))

                for ax,COL in [(axEntr,ENTRCOL)]:
                    ax.plot( dat[:,MUCOL], dat[:,COL], color=color, label='${:0.2f}$'.format(Tval))

                ndat  = dat[:,DENSCOL]
                mudat = dat[:,MUCOL] 
                cdat =  compressibility( mudat, ndat )
                axCompr.plot( mudat, cdat, color=color, label='${:0.2f}$'.format(Tval))
                axComprn.plot( ndat, cdat, color=color, label='${:0.2f}$'.format(Tval))
                
            except Exception as e:
                raise e 
                pass 
            
    
    
        # PLOT THE QMC DATA 
        # data columns 
        MUCOL = 0
        DENSCOL = 1
        SPInCOL = 3
        ENTRCOL = 7
        ms = 1.6; mew=0.3
        
        # get SPI data 
        fname =  '../QMC_Final/'+ 'U{U:02d}/T{T:0.3f}.dat'.format(U=int(Uval),T=Tval)
        if not os.path.isfile(fname):
            #print "File does not exist: ", fname 
            pass
        else:
            try:
                dat = np.loadtxt(fname)

                for ax,COL in [(axn,DENSCOL), (axSn,SPInCOL)  ]:
                    ax.plot( dat[:,MUCOL], dat[:,COL], 'o', \
                            ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))
                axSnn.plot( dat[:,DENSCOL], dat[:,SPInCOL], 'o', \
                            ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))
                
                ndat  = dat[:,DENSCOL]
                mudat = dat[:,MUCOL] 
                cdat =  compressibility( mudat, ndat )
                axCompr.plot( mudat, cdat, 'o', \
                            ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))
                axComprn.plot( ndat, cdat, 'o', \
                            ms=ms, mew=mew, mec=color, mfc='None', \
                            label='${:0.3f}$'.format(Tval))
                
            except:
                pass 
        
        try:
            # get ENTROPY data
            fname =  '../QMC_Final/'+ 'U{U:02d}/S{T:0.3f}.dat'.format(U=int(Uval),T=Tval)
            dat = np.loadtxt(fname)

            for ax,COL in [(axEntr,ENTRCOL)]:
                ax.plot( dat[:,MUCOL], dat[:,COL], 'o', \
                        ms=ms, mew=mew, mec=color, mfc='None',\
                        label='${:0.3f}$'.format(Tval))
        except:
            pass 
        
        
        # PLOT THE QMC_EXTRA DATA
        # data columns 
        MUCOL = 0
        DENSCOL = 1
        SPInCOL = 3
        ENTRCOL = 7
        ms = 1.6; mew=0.3
        try: 
            # get SPI data 
            fname =  '../QMC_Extra_Final/'+ 'U{U:02d}/T{T:0.3f}.dat'.format(U=int(Uval),T=Tval)
            dat = np.loadtxt(fname)

            for ax,COL in [(axn,DENSCOL), (axSn,SPInCOL)  ]:
                ax.plot( dat[:,MUCOL], dat[:,COL], 's', \
                        ms=ms, mew=mew, mec=color, mfc='None', \
                        label='${:0.3f}$'.format(Tval))
            axSnn.plot( dat[:,DENSCOL], dat[:,SPInCOL], 's', \
                        ms=ms, mew=mew, mec=color, mfc='None', \
                        label='${:0.3f}$'.format(Tval))
            
            ndat  = dat[:,DENSCOL]
            mudat = dat[:,MUCOL] 
            cdat =  compressibility( mudat, ndat )
            axCompr.plot( mudat, cdat, 'o', \
                        ms=ms, mew=mew, mec=color, mfc='None', \
                        label='${:0.3f}$'.format(Tval))
            axComprn.plot( mudat, cdat, 'o', \
                        ms=ms, mew=mew, mec=color, mfc='None', \
                        label='${:0.3f}$'.format(Tval))
            
        except:
            pass 
        
        try:
            # get ENTROPY data
            fname =  '../QMC_Extra_Final/'+ 'U{U:02d}/S{T:0.3f}.dat'.format(U=int(Uval),T=Tval)
            dat = np.loadtxt(fname)

            for ax,COL in [(axEntr,ENTRCOL)]:
                ax.plot( dat[:,MUCOL], dat[:,COL], 's', \
                        ms=ms, mew=mew, mec=color, mfc='None',\
                        label='${:0.3f}$'.format(Tval))
        except:
            pass 
        

            
    hl = 0.8; nc=2; cs=0.3
    axn.legend(title='$T/t$',bbox_to_anchor=(1.02,-0.02), loc='lower right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axSn.legend(title='$T/t$',bbox_to_anchor=(1.02,0.99), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axSnn.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs )
    axEntr.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axCompr.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)
    axComprn.legend(title='$T/t$',bbox_to_anchor=(1.02,1.05), loc='upper right', numpoints=1, \
           prop={'size':6}, handlelength=hl, handletextpad=0.5, ncol=nc, columnspacing=cs)

    axn.set_ylim(-0.2, 2.2) 
    axSn.set_ylim(-0.2,12.0)
    axSnn.set_ylim(-0.2,12.0)
    axEntr.set_ylim( 0., 1.2 )
    axCompr.set_ylim(-0.02,0.30)
    axComprn.set_ylim(-0.02,0.30)

    if i != 0 and i != 5 and i != 10 and i!=15 :
            axn.yaxis.set_ticklabels([])
            axSn.yaxis.set_ticklabels([])
            axSnn.yaxis.set_ticklabels([])
            axEntr.yaxis.set_ticklabels([])
            axCompr.yaxis.set_ticklabels([])
            axComprn.yaxis.set_ticklabels([])
    else:
        axn.set_ylabel('$n$', rotation=0,labelpad=11)
        axSn.set_ylabel(r'$\frac{S_{\pi}}{n}$', rotation=0,labelpad=11, fontsize=14)
        axSnn.set_ylabel(r'$\frac{S_{\pi}}{n}$', rotation=0,labelpad=11, fontsize=14)
        axEntr.set_ylabel(r'$s$', rotation=0,labelpad=11, fontsize=14)
        axCompr.set_ylabel(r'$\kappa_{T}/\kappa_{T}^{0}$', rotation=90,labelpad=8, fontsize=14)
        axComprn.set_ylabel(r'$\kappa_{T}/\kappa_{T}^{0}$', rotation=90,labelpad=8, fontsize=14)
        
    if i in range(11):
        axn.xaxis.set_ticklabels([])
        axSn.xaxis.set_ticklabels([])
        axSnn.xaxis.set_ticklabels([])
        axEntr.xaxis.set_ticklabels([])
        axCompr.xaxis.set_ticklabels([])
        axComprn.xaxis.set_ticklabels([])
    else:
        axn.set_xlabel(r'$\mu/t$',labelpad=-1)
        axSn.set_xlabel(r'$\mu/t$',labelpad=-1)
        axSnn.set_xlabel(r'$n$',labelpad=-1)
        axEntr.set_xlabel(r'$\mu/t$',labelpad=-1)
        axCompr.set_xlabel(r'$\mu/t$',labelpad=-1)
        axComprn.set_xlabel(r'$n$',labelpad=-1)

    axn.set_xlim(-10.,36)
    axSn.set_xlim(-10.,36)
    axSnn.set_xlim(0.,2.5)
    axEntr.set_xlim(-10., 36.)
    axCompr.set_xlim(-10., 36.)
    axComprn.set_xlim(-0.2,2.2)
    
    for ax in [axn, axSn, axSnn, axEntr, axCompr, axComprn]:
        ax.grid(alpha=0.4)
        ax.xaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6, prune='lower'))
        ax.text( 0.04, 1.0, '$U/t={:d}$'.format(int(Uval)), \
                 ha='left', va='top', transform=ax.transAxes,
                 bbox={'boxstyle':'round','facecolor':'white'})


name = [\
        r'$\mathrm{Density}$',\
        r'$\mathrm{AFM\ } S_{\pi}\ \mathrm{per\ particle}$',\
        r'$\mathrm{AFM\ } S_{\pi}\ \mathrm{per\ particle}$',\
        r'$\mathrm{Entropy}$',\
        r'$\mathrm{Compressibility}$',\
        r'$\mathrm{Compressibility}$',\
        
        ]
for j,fig in enumerate([fign, figSn, figSnn, figEntr,figCompr, figComprn]):
    fig.text(0.90, 0.16, name[j], fontsize=14, \
             ha='center', va='center', linespacing=1.6)
    

    
fign.savefig('ALL_Final/density_All.png'.format(Tval), dpi=300)
plt.close(fign)
figSn.savefig('ALL_Final/spi_n_All.png'.format(Tval), dpi=300)
plt.close(figSn)
figSnn.savefig('ALL_Final/spi_n_n_All.png'.format(Tval), dpi=300)
plt.close(figSnn)
figEntr.savefig('ALL_Final/entropy_All.png'.format(Tval), dpi=300)
plt.close(figEntr)
figCompr.savefig('ALL_Final/compressibility_All.png'.format(Tval), dpi=300)
plt.close(figCompr)
figComprn.savefig('ALL_Final/compressibility_n_All.png'.format(Tval), dpi=300)
plt.close(figCompr)
```
