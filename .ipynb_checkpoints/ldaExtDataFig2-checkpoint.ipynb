## Very dense sampling of QMC in U-space will be done to extract a smooth curve

```python
import qmc
import os 
Tcommon = [0.7,0.8,1.0, 1.2, 1.4]
Tcommon = [0.69]
Tcommon = [0.75]
Tcommon = [0.90] 
for i,T0 in enumerate( Tcommon):
    Uvals = np.linspace( 80.*11.1/290.,  780.*11.1/290., 180 )
    #Uvals = np.array([200., 245., 290., 335., 380., 425., 470.] ) 
    #Uvals = Uvals * 11.1 / 290. 
    spi = np.empty_like( Uvals )
    for uu,U in enumerate(Uvals):
        #print U 
        spi[uu] = qmc.find_closest_qmc(U=U, T=T0, mu=U/2.)
        
    savedir = 'dataplots/SEXP/QMC_hom/'
    if not os.path.exists(savedir):
        os.makedirs(savedir)
    fname = 'T{:0.2f}_qmc.dat'.format(T0)
    np.savetxt(savedir+fname, np.column_stack((Uvals, spi)))
    plt.plot( Uvals, spi,'.' )
plt.gca().set_xlim(0., 30.)
```

## Smoothing of the QMC dense set above is performed here 

```python


Tcommon = [0.7,0.8,1.0, 1.2, 1.4]
Tcommon = [0.69] 
Tcommon = [0.75]
Tcommon = [0.90]
for i,T0 in enumerate( Tcommon):
    savedir = 'dataplots/SEXP/QMC_hom/'
    fname = 'T{:0.2f}_qmc.dat'.format(T0)
    dat = np.loadtxt( savedir + fname )
    
    x = dat[:,0]
    y = dat[:,1]
    plt.plot( x,y,'.')
    
    z = np.polyfit(x, y, 4)
    p = np.poly1d(z)
    yz = p(x)
    plt.plot( x, yz, lw=3.)
    
    rel_residual = (yz - y)/yz 
    err = np.std(rel_residual)
    print err
    err = yz * err 
    
    plt.fill_between( x,  yz-err, yz+err)
    
    
    fname = 'T{:0.2f}_qmc_smooth.dat'.format(T0)
    np.savetxt( savedir + fname,  np.column_stack(( x, yz, err )))

    

    
    
```
