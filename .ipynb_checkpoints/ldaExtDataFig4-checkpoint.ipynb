```python

from uncertainties import ufloat,unumpy
from scipy import stats

wF = np.loadtxt('dataplots/SEXP/tANDU.dat')
wF = wF[ np.logical_and(  wF[ :, 0] > 6.5 , wF[:,0]<7.5  ) ] 
z = np.polyfit( wF[:,0], np.power( wF[:,2], 1./3.) , 1)  
p = np.poly1d(z) 
#
# get the slope of wF^(1/3)  in units of wF^(1/3) 
slope_wF13 = p[0] / p(7.) 
#print slope_wF13  
#x = np.linspace( 6., 8., 20) 
#plt.plot( wF[:,0], np.power(wF[:,2], 1./3.), '.', color='black')
#plt.plot( x, p(x), '-' )  
#plt.show()

# get the uncertainty in the lattice depth for x,y,z
sx = ufloat( 21.8, 0.3)
sy = ufloat( 21.6, 0.2)
sz = ufloat( 18.6, 0.5)

def get_err_sq( s ):
    mW = 7.0 * s.nominal_value 
    return  ( (mW / sx).std_dev() )**2.
 
s0_err =  np.sqrt(  get_err_sq( sx ) \
             + get_err_sq( sy ) + get_err_sq( sz ) )

U_err_factor = slope_wF13 * s0_err 

#print slope_wF13
#print get_err_sq( sx ) 
#print get_err_sq( sy ) 
#print get_err_sq( sz ) 
#print  s0_err 
#print U_err_factor


# Start matplotlib
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

sc = 1.1
fig = plt.figure(figsize=(4.8*sc,2.8*sc))
gs = matplotlib.gridspec.GridSpec( 1,1, wspace=0.25, hspace=0.11,\
                    top=0.85, left=0.1, right=0.53, bottom=0.16)

ax = fig.add_subplot( gs[0] )  
ax.set_ylim(0.6, 2.2 ) 

from scipy.interpolate import interp1d

def fx(x):
    return  x * 11.1 / 290. 

def plotexp( ax, dat, mfc, lstr, **kwargs):
    X=fx(dat[:,0]); Y=dat[:,3];YERR=dat[:,4]
    XERR = X * U_err_factor / 2. 
    marker = 'o'; mec='black'
    ax.errorbar( X, Y, yerr=YERR, xerr=XERR, \
            fmt='.', capsize=1.2, elinewidth=0.7, mew=0.5, marker=marker,\
            mfc=mfc, mec=mec, ms=3.7, ecolor=mec, **kwargs )
    ax.plot( X, Y,  '.', marker=marker, ms=3.7, mfc=mfc, mec=mec, label=lstr, **kwargs)

    
datSpi = np.loadtxt('dataplots/SEXP/experiment_data_Spi.txt')
datSth = np.loadtxt('dataplots/SEXP/experiment_data_Sth.txt')
for dat,mfc,lstr in [(datSpi,'black',r'$\mathbf{Q}=\mathbf{\pi}$' ), (datSth,'white', r'$\mathbf{Q}=\mathbf{\theta}$' )]:
    plotexp( ax, dat, mfc, lstr)


# YLABELS 
ax.set_ylabel( r'$S_{\mathbf{Q}}$', rotation=0, fontsize=13, labelpad=12. )
ax.set_xlabel( r'$U_{0}/t_{0}$', fontsize=13) 
    
# AXES GENERAL PROPERTIES 
ax.grid(alpha=0.3)
ax.axhline( 1.0, ls='-', color='0.7', lw=1.2, zorder=0.1)
ax.set_xlim(0., 25.)

ax.legend(loc='best',numpoints=1,prop={'size':9},\
    handlelength=1.1,handletextpad=0.5)
ax.yaxis.set_major_locator( \
    matplotlib.ticker.MultipleLocator( 0.2 ))
#ax.yaxis.set_minor_locator( \
#    matplotlib.ticker.MultipleLocator( 0.2 ))
ax.xaxis.set_major_locator( \
    matplotlib.ticker.MaxNLocator( nbins=6))



def ascatt(x):
    return x * 290. / 11.1
def Ut(x):
    return x * 11.1 / 290. 

aSs =  [ 80., 200., 290., 380., 470, 560. ] 
ax2 = ax.twiny()
ax2.set_xticks( [ Ut(xi) for xi in  aSs ] )  
ax2.set_xticklabels( [ r'$%d$' % xi for xi in aSs ] )
ax2.tick_params(axis='x', labelsize=10)
ax2.set_xlim( 0. , 25. )
ax2.set_xlabel('$\mathrm{Scattering\ length}\ (a_{0})$', fontsize=10, labelpad=7.5) 


#### PLOT THE QMC COMPARISON INCLUDING LDA 

import glob
from colorChooser import rgb_to_hex, cmapCycle

gsLDA = matplotlib.gridspec.GridSpec(2,1,\
            wspace=0.1, hspace=0.34,\
            left=0.64, right=0.90, bottom=0.12, top=0.98)
axLDA = fig.add_subplot( gsLDA[0] )
axLDA.set_ylabel(r'$S_{\pi}$', rotation=0, labelpad=10., fontsize=10 )
axLDA.set_xlim(1., 24.)
axLDA.set_xlabel('$U_{0}/t_{0}$', fontsize=10, labelpad=0.8)
axLDA.set_ylim(0.8, 2.2 ) 
axLDA.yaxis.set_major_locator( matplotlib.ticker.MultipleLocator(0.4))
axLDA.yaxis.set_minor_locator( matplotlib.ticker.MultipleLocator(0.2))
axLDA.grid(alpha=0.25, which='both')
axLDA.axhline( 1.0, ls='-', color='0.7', lw=1.2, zorder=0.1)

axTH = fig.add_subplot( gsLDA[1] )
axTH.set_xlabel(r'$T/t_{*}$', fontsize=10, labelpad=0.8)
axTH.set_ylabel(r'$S_{\mathbf{\pi}}$', rotation=0., labelpad=10., fontsize=10)



aSs = [200, 290, 380, 470]
Tsets = []
for aS in aSs:
    fdats  = sorted(glob.glob('dataplots/SPI/{:03d}/*.dat'.format(int(aS))))
    Tspis  = [ float(f.split('Tspi')[1].split('.dat')[0]) for f in fdats ] 
    print aS
    print Tspis
    Tsets.append( set(Tspis))
    
    
Tcommon = sorted(set.intersection(*Tsets))
print "Common:"
print Tcommon


resultsThermometer = [] 

for i,Tspi in enumerate(Tcommon):
    
    #color = cmapCycle( matplotlib.cm.bone, float(i+3), lbound=0, ubound=float(len(Tcommon)+4))
    #color = cmapCycle( matplotlib.cm.winter, float(i), lbound=0, ubound=float(len(Tcommon)+0))
    color = cmapCycle( matplotlib.cm.cool, float(i), lbound=0, ubound=float(len(Tcommon)+0), inv=False)
      
    if Tspi == 0.50:
        continue
    result = []
    # Columns are :  N/1e5,  SpiBulk, Frac, S/N_Bulk, t*, T*, U*
    for j,(aS,U,bestN) in enumerate([\
            (200,7,4), \
            (245,9,0), \
            (290,11,3), \
            (335,12,0), \
            (380,14,2), \
            (425,16,0), \
            (470,17,3)\
            ]):
        try:
            fname = 'dataplots/SPI/{a:03d}/aS{a:03d}_U{U:02d}_Tspi{T:0.2f}.dat'.format(a=aS,U=U,T=Tspi)
            dat = np.loadtxt( fname )
            dat = np.atleast_2d( dat ) 
            # 0 is number
            # 1 is Spi Bulk
            # 3 is Entropy Bulk
            # 4 is t^{*}
            # 5 is T/t^{*} 
            # 6 is U^{*} / t^{*}
            result.append( [ fx(aS), dat[bestN][1], dat[bestN][0], dat[bestN][3] , dat[bestN][5], dat[bestN][6] ] ) 
            
            print "U0/t0 = {:0.2f}  T/t0 = {:0.2f},  T/t* = {:0.2f}".format( U, Tspi, dat[bestN][5])
            
            if aS == 290:
                resultsThermometer.append( [  dat[bestN][5], dat[bestN][1] ]  )
        except Exception as e:
            print "Did not find aS={:03d}, T={:0.2f}".format(aS,Tspi)
            print e
            pass
    result = np.array(result)
    print "U/t star  =",  result[:,5]
    #print "mean Tstar = ", np.mean( result[:,4] ) 
    #print " std Tstar = ", np.std( result[:,4] )
    #print result 
    
    # Choose what goes on the legend title
    Tlabel = np.mean(result[:,4] )
    #~Tlabel = Tspi
    
    axLDA.plot( result[:,0], result[:,1], 'D-', ms=2.5, lw=1.4, color=color,\
               mfc='white', mec=color, label='{:0.2f}'.format(Tlabel))



axTH.axhspan( 1.962-0.115, 1.962+0.115, facecolor='gray', alpha=0.6)
resultsThermometer = np.array( resultsThermometer )
x = resultsThermometer[:,0]
y = resultsThermometer[:,1]
axTH.plot( x, y, 'D', mfc='white', mec='black', ms=3.5, mew=1., color='black')
axTH.set_xlim(0.43, 0.93)
axTH.set_ylim(1.2, 2.38)
axTH.grid(alpha=0.3)


for tick in axLDA.get_xaxis().get_major_ticks():
    tick.set_pad(2.5)
for tick in axTH.get_xaxis().get_major_ticks():
    tick.set_pad(2.5)
    
axTH.text(1.02, 0.975, '$U_{0}/t_{0}=11.1$',\
          ha='right', va='top', transform=axTH.transAxes,
          bbox={'boxstyle':'round', 'facecolor':'white'})

    
expdat = np.loadtxt('dataplots/SEXP/experiment_data_Spi.txt')
plotexp( axLDA, expdat, 'black',r'$\mathbf{\pi}$', zorder=10.)
#l0 = axLDA.legend(bbox_to_anchor=(0.0,1.0), loc='upper left', numpoints=1, \
#            prop={'size':10}, handlelength=1., handletextpad=0.5)
#axLDA.add_artist(l0)

handles, labels = axLDA.get_legend_handles_labels()
legend = axLDA.legend(handles[:-1], labels[:-1], title='$T/t_{*}$',bbox_to_anchor=(1.38,1.05), loc='upper right', numpoints=1, \
    prop={'size':8}, handlelength=0.7, handletextpad=0.5)
plt.setp( legend.get_title(), fontsize=9)




ax.text(0.08,0.95, r'\textbf{a}', ha='left', va='top', transform=ax.transAxes,\
         fontsize=12.3)
axLDA.text(0.08,0.95, r'\textbf{b}', ha='left', va='top', transform=axLDA.transAxes,\
         fontsize=12.3)
axTH.text(0.08,0.08, r'\textbf{c}', ha='left', va='bottom', transform=axTH.transAxes,\
         fontsize=12.3)
    
    
import datetime
datestamp = datetime.datetime.now().strftime('%Y%m%d_%H%M%S_')
datestamp = ''

#fig.savefig('dataplots/SEXP/' + datestamp+'udat_final_onerow_XU.png', dpi=350)
fig.savefig('dataplots/SEXP/04_SpiLDA_.png', dpi=350)
```
